```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Пусть есть ф-я
def func(x):
    return x ** 3 - 6 * x ** 2 - 15 * x + 10
```

```python
x = np.linspace(-5, 10, 150)
```

```python
fig = plt.figure()
fig.set_size_inches(10,8)
plt.plot(x, func(x))
plt.ylabel('f')
plt.xlabel('x')
plt.show()
```
